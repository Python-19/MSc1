# Data Visualization

```python
import pandas as pd
wisconsin = pd.read_csv('F:\FDS\dataset\wisconsin.csv', index_col = 0)
wisconsin.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
data = wisconsin['texture_mean'].value_counts()
points = data.index
frequency = data.values
ax.bar(points, frequency)
ax.set_title('Wisconsin')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
cars_data = pd.read_csv('F:\FDS\dataset\wisconsin.csv', index_col = 0)
cars_data.head()
```

```python
sns.set(style = 'darkgrid')
sns.regplot(x = cars_data['texture_mean'], y = cars_data['perimeter_mean'])
```

```python
sns.lmplot(x = 'texture_mean', y = 'area_mean', data = cars_data, fit_reg = False, hue = 'diagnosis',
           legend = True, palette = 'Set1')
```

```python
sns.distplot(cars_data['area_mean'])
```

```python
sns.distplot(cars_data['area_mean'], kde = False, bins = 5)
```
